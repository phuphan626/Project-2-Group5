# Pitching Averages Cleanup

```python
# Import Dependencies
import pandas as pd
from config import my_key
import json
```

```python
# Set the main url, then read in the table. There are two tables in the web page. We picked the Pitching Averages table.
# The first one is "League Year-By-Year Pitching--Averages"
baseball_url='https://www.baseball-reference.com/leagues/MLB/pitch.shtml'
pitching_average_table=pd.read_html(baseball_url)[0]
# The second one is "League Year-By-Year Pitching--Totals"
pitching_total_table=pd.read_html(baseball_url)[1]
```

```python
# Convert it to a dataframe 
pitching_average_df=pd.DataFrame(pitching_average_table)
```

```python
# Set the index to Year column for later cleaning
pitching_average_df=pitching_average_df.set_index('Year')
```

```python
# Display the table
pitching_average_df
```

## We notice that when pandas reads in the table, there are some duplicate rows that do not contain the decimal value. Therefore, we want to drop them out of the table. Also the table has 32 columns, which is redundant, so we filter out some columns, and keep the necessary ones.

```python
# Filter out the duplicate headers as rows
rows_to_drop=pitching_average_df.loc['Year']
```

```python
# Display to check
rows_to_drop
```

```python
# Drop out those unnecessary rows
cleaned_baseball_df=pitching_average_df[pitching_average_df['Tms'].isin(rows_to_drop)==False]
```

```python
# Checking the length of the df. The original has 156 rows, so the cleaned up one should have 150 rows after dropping.
len(cleaned_baseball_df)
```

```python
# Dropping those unnecessary columns
cleaned_baseball_df=cleaned_baseball_df[['Tms','R/G','ERA','GF','CG','HR','SO','HBP','HR9','SO9']]
```

```python
# Reset index to Year column
cleaned_baseball_df=cleaned_baseball_df.reset_index()
```

```python
# Convert the Year to integer
cleaned_baseball_df['Year']=cleaned_baseball_df['Year'].astype(int)
```

```python
# Get data from 1920 and up.
cleaned_baseball_df=cleaned_baseball_df.loc[(cleaned_baseball_df['Year']>=1920)]
```

```python
# Convert the rest of columns to float type, so we can load them into PostgreSQL
cleaned_baseball_df[['Tms','R/G','ERA','GF','CG','HR','SO','HBP','HR9','SO9']]=cleaned_baseball_df[['Tms','R/G','ERA','GF','CG','HR','SO','HBP','HR9','SO9']].astype(float)
```

```python
# Display the final table
cleaned_baseball_df
```

```python
# Export to html format to use in the index.html
cleaned_baseball_df.to_html('html_files/pitching.html')
```

```python
# Export wo json format for visualization
cleaned_baseball_df.to_json('json_files/pitching.json',orient='records')
```

# Batting Cleanup

```python
# Doing the cleanup for batting stats
batting_url = "https://www.baseball-reference.com/leagues/MLB/bat.shtml"
batting_table = pd.read_html(batting_url)[0]
```

```python
# Convert it to DataFrame
batting_df = pd.DataFrame(batting_table)
batting_df
```

```python
# Setting the index to Year column
batting_df = batting_df.set_index('Year')
```

```python
# Filter out unnecessary rows as above
drop_rows = batting_df.loc['Year']
drop_rows
```

```python
# Drop those rows
batting_cleaned_df = batting_df[batting_df['Tms'].isin(drop_rows)==False]
```

```python
# Checking the length of the cleaned table
len(batting_cleaned_df)
```

```python
# Fill the NaN cells with 0
batting_cleaned_df = batting_cleaned_df.fillna(0)
batting_cleaned_df
```

```python
# Filter out desired columns
batting_cleaned_df = batting_cleaned_df[['Tms', '2B', '3B', 'SB', 'BA', 'OPS']]
batting_cleaned_df
```

```python
# Reset the index
batting_cleaned_df = batting_cleaned_df.reset_index()
```

```python
# Convert the Year column to integer type
batting_cleaned_df['Year'] = batting_cleaned_df['Year'].astype(int)
```

```python
# Convert the rest to float type
batting_cleaned_df[['Tms', '2B', '3B', 'SB', 'BA', 'OPS']]=batting_cleaned_df[['Tms', '2B', '3B', 'SB', 'BA', 'OPS']].astype(float)
```

```python
# Filter out cells before 1920
cleaned_date = batting_cleaned_df.loc[(batting_cleaned_df['Year']>=1920)]
cleaned_date
```

```python
# Export to html format
cleaned_date.to_html('html_files/batting.html')
```

```python
# Export to json format
cleaned_date.to_json('json_files/batting.json',orient='records')
```
