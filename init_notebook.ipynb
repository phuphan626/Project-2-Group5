```python
# Import Dependencies
import pandas as pd
```

```python
# Set the main url, then read in the table. There are two tables in the web page
# The first one is "League Year-By-Year Pitching--Averages"
baseball_url='https://www.baseball-reference.com/leagues/MLB/pitch.shtml'
baseball_average_table=pd.read_html(baseball_url)[0]
# The second one is "League Year-By-Year Pitching--Totals"
baseball_total_table=pd.read_html(baseball_url)[1]
```

```python
# Convert to a dataframe
baseball_totals_df=pd.DataFrame(baseball_total_table)
```

```python
# Set the index to Year column
baseball_totals_df=baseball_totals_df.set_index('Year')
```

```python
# Convert it to a dataframe 
baseball_average_df=pd.DataFrame(baseball_average_table)
```

```python
# Set the index to Year column for later cleaning
baseball_average_df=baseball_average_df.set_index('Year')
```

```python
# Filter out the duplicate headers as rows
rows_to_drop=baseball_average_df.loc['Year']
```

```python
# Filter out the duplicate headers as rows
totals_df_rows=baseball_totals_df.loc['Year']
```

```python
# Display to check
totals_df_rows
```

```python
# Drop out those unnecessary rows
cleaned_baseball_df=baseball_average_df[baseball_average_df['Tms'].isin(rows_to_drop)==False]
```

```python
# Checking the length of the df. The original has 156 rows, so the cleaned up one should have 150 rows after dropping.
len(cleaned_baseball_df)
```

```python
# Drop out those unnecessary rows
cleaned_totals_df=baseball_totals_df[baseball_totals_df['Tms'].isin(totals_df_rows)==False]
```

```python
# Checking for the length
len(cleaned_totals_df)
```

```python
# Fill NaN for all the cells in two tables
cleaned_totals_df=cleaned_totals_df.fillna(0)
cleaned_baseball_df=cleaned_baseball_df.fillna(0)
```

```python
# Display to check
cleaned_totals_df['ERA+'].unique()
```

```python
# Dropping those unnecessary columns
cleaned_baseball_df=cleaned_baseball_df[['Tms','R/G','ERA','GF','CG','HR','SO','HBP','HR9','SO9']]
```

```python
cleaned_baseball_df=cleaned_baseball_df.reset_index()
```
